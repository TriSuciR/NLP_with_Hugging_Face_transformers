```python
from transformers import pipeline

# Membuat pipeline untuk klasifikasi zero-shot
classifier = pipeline("zero-shot-classification")

# Kalimat yang akan diklasifikasi
text = "Ini adalah panduan lengkap tentang cara memulai bisnis digital"

# Label kandidat untuk klasifikasi
candidate_labels = ["pendidikan", "bisnis", "kesehatan"]

# Menjalankan klasifikasi zero-shot
classifier(text, candidate_labels=candidate_labels)
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
text = "Banyak orang mulai berinvestasi di aset kripto"
candidate_labels = ["teknologi", "ekonomi", "investasi"]
classifier(text, candidate_labels=candidate_labels)
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
text = "Banyak perusahaan sekarang menggunakan kecerdasan buatan untuk meningkatkan efisiensi kerja"
candidate_labels = ["teknologi", "bisnis", "otomasi"]
classifier(text, candidate_labels=candidate_labels)
```

```python
from transformers import pipeline

generator = pipeline("text-generation", model="distilgpt2")

text = "Di dalam seminar ini, kita akan membahas cara-cara"

results = generator(
    text,
    max_length=30,
    num_return_sequences=2,
)

for i, result in enumerate(results):
    print(f"Generated Sequence {i + 1}: {result['generated_text']}")
```

```python
from transformers import pipeline

generator = pipeline("text-generation", model="distilgpt2")

text = "Pada sesi pelatihan ini, kita akan membahas cara menggunakan"

results = generator(
    text,
    max_length=30,
    num_return_sequences=2,
)
for i, result in enumerate(results):
    print(f"Teks yang dihasilkan {i+1}: {result['generated_text']}")
```

```python
from transformers import pipeline

unmasker = pipeline("fill-mask")

text = "Pelatihan ini akan memperkenalkan Anda pada <mask> teknologi."

results = unmasker(text, top_k=2)

for i, result in enumerate(results):
    print(f"Hasil {i+1}: {result['sequence']}")
```

```python
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)
text = "Nama saya Andi dan saya belajar di Universitas Indonesia di Jakarta."
results = ner(text)
for entity in results:
    print(entity)
```

```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
context = "Bumi adalah planet ketiga dari Matahari dan merupakan satu-satunya planet yang diketahui memiliki kehidupan."
question = "Apa itu Bumi?"
result = qa_pipeline(question=question, context=context)
print(f"Jawaban: {result['answer']}")
```

```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
context = "Albert Einstein (14 Maret 1879 – 18 April 1955) adalah seorang fisikawan teoretis Jerman yang terkenal dengan teori relativitasnya, khususnya persamaan E=mc². Ia dianggap sebagai salah satu ilmuwan paling berpengaruh dalam sejarah."
question = "Siapa Albert Einstein?"
result = qa_pipeline(question=question, context=context)
print(f"Jawaban: {result['answer']}")
```

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
text = "Makanan di restoran ini sangat enak! Saya pasti akan datang lagi."
result = classifier(text)
print(result)
```

```python
from transformers import pipeline

summarizer = pipeline("summarization")

text = """
Pembangunan infrastruktur di Indonesia mengalami kemajuan pesat dalam beberapa tahun terakhir.
Proyek-proyek besar seperti pembangunan jalan tol, jembatan, dan bandara baru telah
mempercepat aksesibilitas antar daerah. Meskipun demikian, masih ada tantangan dalam
mengatasi masalah lingkungan dan dampak sosial dari pembangunan tersebut.
Pemerintah berupaya untuk menjaga keseimbangan antara pembangunan ekonomi dan
pelestarian lingkungan dengan menerapkan kebijakan yang lebih berkelanjutan.
"""

summary = summarizer(text)

print(summary[0]['summary_text'])
```

```python
from transformers import pipeline

translator = pipeline("translation", model="Helsinki-NLP/opus-mt-id-en")
text_to_translate = "Saya senang berolahraga di pagi hari."
result = translator(text_to_translate)
print(result[0]['translation_text'])
```

# Analisis




1. Zero-Shot Classification, Kode ini menggunakan klasifikasi zero-shot untuk mengkategorikan teks tanpa pelatihan khusus. Fleksibel dan berguna untuk situasi di mana label tidak pasti, cocok untuk aplikasi pendidikan.

2. Text Generation, Menggunakan model generatif untuk menyelesaikan kalimat berdasarkan konteks yang diberikan. Berguna dalam pembuatan konten dan dialog, menunjukkan bagaimana model dapat menghasilkan teks yang berkelanjutan.

3. Named Entity Recognition (NER), Mengimplementasikan NER untuk mengidentifikasi entitas penting dalam teks, seperti nama dan lokasi. Sangat penting untuk ekstraksi informasi dan analisis data dalam aplikasi seperti chatbot.

4. Question Answering, Kode ini membangun sistem tanya jawab yang memberikan jawaban berdasarkan konteks. Berguna untuk pencarian informasi dan aplikasi edukasi, memungkinkan pengguna mendapatkan jawaban langsung dan akurat.

5. Sentiment Analysis, Menganalisis sentimen untuk menentukan emosi dalam teks. Ini membantu dalam memahami umpan balik pelanggan dan menilai kepuasan, sangat berharga untuk bisnis.

6. Summarization, Model digunakan untuk merangkum teks panjang menjadi ringkasan yang lebih pendek. Membantu pengguna mendapatkan informasi inti dengan cepat, relevan dalam dunia yang penuh informasi.

7. Translation, Menggunakan model terjemahan untuk menerjemahkan teks dari satu bahasa ke bahasa lain. Memfasilitasi komunikasi lintas budaya dan meningkatkan aksesibilitas informasi.
